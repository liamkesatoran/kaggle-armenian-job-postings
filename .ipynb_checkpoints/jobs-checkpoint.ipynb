This notebook exposes some quick and dirty analysis on an Armenian data job posts database found on kaggle : https://www.kaggle.com/madhab/jobposts

This notebook can :

--- Look at the frequency of IT jobs over time + show time frequency correlation

--- Look how many job listings require a certain keyword (like 'English' or 'Python")

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df= pd.read_csv('data job posts.csv')
df
```

The date column entries are string type, so I won't use it. I'll use month and year columns instead.

```python
df['Year']
```

```python
df['Month']
```

Converting Year + Month into a single 'int_date' that increments by 1 for each month

```python
df['int_date']= df['Year']*12 + df['Month'] -1
df['int_date'] += -np.min(df['int_date'])
df['int_date']
```

int_date 0 is now Jan 2004 and int_date 144 is now Dec 2015

```python
df['int_date'].hist(bins=144)
```

Let's check if dates are sorted (they look like it)

```python
np.all(df['int_date'][1:].values >= df['int_date'][:-1].values)
```

They're actually not sorted

```python
df = df.sort_values('int_date')
np.all(df['int_date'][1:].values >= df['int_date'][:-1].values)
```

Now they are

IT columns apparently tells if a job is in IT or not, let's plot its density distribution over time

```python
IT_per_month = df.groupby('int_date').mean()['IT']
plt.plot(IT_per_month)
```

The per-month time series is pretty noisy, let's try grouping by year?

```python
IT_per_year = df.groupby('Year').mean()['IT']
plt.plot(IT_per_year)
```

is the 2008-2009 dip due to the financial crisis ?

pearsonr returns R coefficient  + a p- value, see:

https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.pearsonr.html

```python
from scipy.stats.stats import pearsonr 
pearsonr(IT_per_year.index, IT_per_year.values)
```

```python
pearsonr(IT_per_month.index, IT_per_month.values)
```

There is a positive correlation between date and frequency of IT jobs, "meaning" that  there are more IT jobs (frequency wise) as time passed from 2004 to 2015. Which makes sense to me...


There is less correlation for the per-month time series rather than the per-year (noisier ?) but lower p-value too (meaning the chance of a sample of same size from a random distribution having at least such a correlation coefficient is low, so we are pretty sure the correlation is not a random artifact

Let's search for specific words in the required qualifications

```python
def search_for(string,sentence):
    try:
        ans = string in sentence
    except :
        ans = False
    return ans
```

We can now look for specific keywords, for instance Linux :

```python
df['RequiredQual'].apply(lambda x : search_for('Linux',x)).sum()
```

Doing things in bulk:

```python
Skills = ['Linux', 'Office', 'Python', 'Ruby', 'HTML', 'PHP', 'C++', 'SQL', 'Java', 'C#', 'Swift']
Pearson_R_vect = []

for skill in Skills:
    needs_skill = df['RequiredQual'].apply(lambda x : search_for(skill,x))
    df[skill]= needs_skill
    pearson= pearsonr(df.groupby('int_date').mean()[skill].index, df.groupby('int_date').mean()[skill].values)
    Pearson_R_vect +=[pearson[0]]
    print(skill,'has a total of', needs_skill.sum(),'job postings',\
          'with a time-series pearson R',pearson[0])
```

Ranking the skills according to R scores ("hottest skills" ?):

Argsort actually sorts from low to high so we need to reverse it

```python
i = 1
for x in np.argsort(Pearson_R_vect)[::-1]:
    print('number', i, 'is', Skills[x])
    i+=1
```
