```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df= pd.read_csv('data job posts.csv')
df
```

The dates are string type, I'll try convert them to a number

```python
str.split(df['date'][1])
```

```python
df['split_date']= df['date'].apply(str.split)
df['split_date']
```

```python
def not_len_3(x):
    return len(x) !=3

bad_df_date =  df['split_date'].apply(not_len_3)
bad_df_date.any()
```

Oops some dates are badly formatted, let's see how many

```python
bad_df_indexes = bad_df_date[bad_df_date].index.values
len(bad_df_indexes)
```

Pretty sizable chunk (2297/19001) , let's see if they are salvagable

```python
df['date'][bad_df_indexes]
```

Missing Year is pretty bad for what I want to do so I'll remove these entries

```python
filtered_df = df[bad_df_date.apply(lambda x :not x)]
filtered_df['split_date'][:][0]
```

```python

months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
dic = dict(zip(months, list(range(len(months)))))
def str_arr_to_int(x):
    month, day, year = x
    date = 12*int(year) + dic[month]
    return date
```

```python
filtered_df['int_date']=  filtered_df['split_date'].apply(str_arr_to_int) -\
            min(filtered_df['split_date'].apply(str_arr_to_int))

filtered_df['int_date']
```
